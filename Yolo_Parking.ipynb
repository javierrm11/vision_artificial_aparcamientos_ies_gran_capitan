```python
!pip install ultralytics
import os

# Importar las librerías necesarias
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# 1. Cargar el modelo YOLOv8 más ligero (versión nano)
model = YOLO('yolov8n.pt')
```

```python
# 2. Ruta exacta a tu video en Drive
ruta_video_drive = '/content/drive/MyDrive/datosejercicios/video_parking.mp4'

# --- VERIFICACIÓN DE SEGURIDAD ---
if not os.path.exists(ruta_video_drive):
    print(f"No encuentro el archivo en: {ruta_video_drive}")
    print("Asegúrate de que el nombre sea exactamente (ojo con las mayúsculas).")
else:
    print(f"Vídeo localizado. Iniciando procesamiento...")

    # 3. Ejecutar la inferencia
    # save=True: guarda el vídeo resultante
    # project y name: crean una carpeta específica para encontrarlo fácil
    model.predict(
        source=ruta_video_drive,
        save=True,
        conf=0.25,
        project='resultados_practica',
        name='video_parking',
        exist_ok=True
    )

    print("\n" + "="*30)
    print("✅ ¡PROCESAMIENTO FINALIZADO!")
    print("El vídeo con las detecciones está en la carpeta:")
    print("📂 resultados_practica/video_coches/video_ejemplo_resultado_personas.avi") # YOLO suele guardarlo como .avi o .mp4
    print("="*30)
```
